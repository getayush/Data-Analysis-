```python
import pandas as pd

# Load the dataset
file_path = r'E:\Case Study 2 data from case study (1).xlsx'
sheet_name = 'Sample_Data_on_Past_Sales_Leads'
df = pd.read_excel(file_path, sheet_name)

# Count of products under each Lead Conversion Class
class_counts = df['Leads_Conversion_Class'].value_counts()

# Display count of products under each class
print("\n\033[1mLead Conversion Class Counts: \033[0m")
print(class_counts)

# Create a pivot table to show which product is sold in which region and its reporting status
pivot_table = pd.pivot_table(df, values='Product', index=['Region', 'Reporting_Status'], columns='Leads_Conversion_Class', aggfunc='count', fill_value=0)

# Display the pivot table
print("\n\033[1mProduct Distribution Across Regions and Reporting Status: \033[0m")
print(pivot_table)
```

```python
import pandas as pd

# Load the dataset
file_path = r'E:\Case Study 2 data from case study (1).xlsx'
sheet_name = 'Sample_Data_on_Past_Sales_Leads'
df = pd.read_excel(file_path, sheet_name)

# Create a pivot table to show which product is sold in which region and its reporting status
pivot_table = pd.pivot_table(df, values='Product', index=['Region', 'Reporting_Status'], columns='Leads_Conversion_Class', aggfunc='count', fill_value=0)

# Add a row for the total number of deals in each region
pivot_table['Total Deals'] = pivot_table.sum(axis=1)

# Add a column for the total number of deals in each Lead Conversion Class
pivot_table.loc['Total Deals'] = pivot_table.sum()

# Display the pivot table
print("\nProduct Distribution Across Regions and Reporting Status:")
print(pivot_table)
```

```python
import pandas as pd

# Load the dataset
file_path = r'E:\Case Study 2 data from case study (1).xlsx'
sheet_name = 'Sample_Data_on_Past_Sales_Leads'
df = pd.read_excel(file_path, sheet_name)

# Create a pivot table to show the distribution of products by Region and Lead Conversion Class
pivot_table_products_by_region = pd.pivot_table(df, values=['Product'], index=['Region', 'Leads_Conversion_Class'], aggfunc='count', fill_value=0, margins=True, margins_name='Total')

# Display the pivot table with totals for each region and won deals for each class in each region
print("\nProduct Distribution by Region and Lead Conversion Class:")
print(pivot_table_products_by_region)

# Calculate and display the total for each region
total_by_region = pd.pivot_table(df, values=['Product'], index=['Region'], aggfunc='count', fill_value=0)
print("\nTotal for Each Region:")
print(total_by_region)

# Calculate and display the number of won deals for each class in each region
won_deals_by_class = pd.pivot_table(df[df['Reporting_Status'] == 'Won'], values=['Product'], index=['Region', 'Leads_Conversion_Class'], aggfunc='count', fill_value=0, margins=True, margins_name='Total')
print("\nNumber of Won Deals by Class in Each Region:")
print(won_deals_by_class)
```

```python
import pandas as pd

# Load the dataset
file_path = r'E:\Case Study 2 data from case study (1).xlsx'
sheet_name = 'Sample_Data_on_Past_Sales_Leads'
df = pd.read_excel(file_path, sheet_name)

# Create a pivot table to show the distribution of products by Industry and Lead Conversion Class
pivot_table_products_by_industry = pd.pivot_table(df, values=['Product'], index=['Industry', 'Leads_Conversion_Class'], columns=['Reporting_Status'], aggfunc='count', fill_value=0, margins=True, margins_name='Total')

# Display the pivot table with totals for each industry and won/lost deals for each class in each industry
print("\nProduct Distribution by Industry, Lead Conversion Class, and Reporting Status:")
print(pivot_table_products_by_industry)

# Calculate and display the total for each industry
total_by_industry = pd.pivot_table(df, values=['Product'], index=['Industry'], columns=['Reporting_Status'], aggfunc='count', fill_value=0, margins=True, margins_name='Total')
print("\nTotal for Each Industry:")
print(total_by_industry)

# Calculate and display the number of won deals for each class in each industry
won_deals_by_class_industry = pd.pivot_table(df[df['Reporting_Status'] == 'Won'], values=['Product'], index=['Industry', 'Leads_Conversion_Class'], aggfunc='count', fill_value=0, margins=True, margins_name='Total')
print("\nNumber of Won Deals by Class in Each Industry:")
print(won_deals_by_class_industry)
```
