```python
df.drop(['description',
 'neighbourhood_group_cleansed',
 'bathrooms',
 'calendar_updated',
 'license'], axis=1, inplace=True)
```

<span style="color:blue">Dropping columns which are completely null </span>

```python
numeric=df.select_dtypes(['int64', 'float64']).columns.tolist()
numeric
```

<span style="color:blue">Identifying the numeric columns</span>

```python
cols=['host_listings_count','host_total_listings_count','bedrooms', 'beds',
      'minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights', 'maximum_maximum_nights',
      'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 
      'review_scores_rating','review_scores_accuracy', 'review_scores_cleanliness','review_scores_checkin',                           
      'review_scores_communication', 'review_scores_location', 'review_scores_value','reviews_per_month']

for column in cols:
    if column in df.columns:
        df[column].fillna(df[column].mean(), inplace=True)
```

<span style="color:blue">Filling the null entries with mean value</span>

```python
df[numeric].isnull().sum()
```

```python
y=df.columns[df.isnull().any()].tolist()
y
```

<span style="color:blue">Finding the other(non-integers) null columns</span>

```python
COLUMNS=['neighborhood_overview','picture_url','host_name','host_since',
         'host_location','host_about','host_response_time','host_response_rate',
         'host_acceptance_rate','host_is_superhost','host_thumbnail_url','host_picture_url',
         'host_neighbourhood','host_has_profile_pic','host_identity_verified',
         'neighbourhood','bathrooms_text','price','has_availability','first_review','last_review']

subset=df[COLUMNS]
subset.head(10)
```

<span style="color:blue">Viewing all categorical columns as a dataframe</span>

```python
COLUMNS=['neighborhood_overview','picture_url','host_name','host_since',
         'host_location','host_about','host_response_time','host_response_rate',
         'host_acceptance_rate','host_is_superhost','host_thumbnail_url','host_picture_url',
         'host_neighbourhood','host_has_profile_pic','host_identity_verified',
         'neighbourhood','bathrooms_text','price','has_availability','first_review','last_review']
for column in COLUMNS:
    df[column].fillna(method='ffill', inplace=True)
```

<span style="color:blue">Forward Filling all the NaN values in the categorical columns</span>

```python
y=df.columns[df.isnull().any()].tolist()
y
```

```python
df['host_neighbourhood'].fillna(method='bfill', inplace=True)
```

<span style="color:blue">Using backward fill for columns 'host_neighbourhood'</span>

```python
y=df.columns[df.isnull().any()].tolist()
y
```

<span style="color:blue">No column is null now</span>

```python
df.drop(['picture_url'], axis=1, inplace=True)
```

```python
ROUND=['bedrooms', 'beds', 'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value', 'reviews_per_month'
]

def round_element(x):
    return round(x,2)

df[ROUND]=df[ROUND].applymap(round_element)
df[ROUND]
```
