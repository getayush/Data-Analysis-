```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

df=df.drop(['FIPS', 'Admin2'], axis=1)
df.head(5)
```

    Dropping FIPS and Admin2 columns and remember to keep axis=1 which means you're dropping columns otherwise if not mentioned     then by default axis=0 which means dropping rows.

```python
df=df.dropna(subset=['Province_State'])
df.head(15)
```

    Removes all rows where Province_State == null

```python
imputer=SimpleImputer(strategy="mean")
df['Lat']=imputer.fit_transform(df[['Lat']])
df['Long_']=imputer.fit_transform(df[['Long_']])
df.head(40)
```

    Impute in Python is a basic strategy to fill missing values with either mean, median, constant or most_frequent value.  

    Estimators are functions in ETL Pipeline which has fit, transform and fit_transform

```python
from sklearn.base import BaseEstimator, TransformerMixin

class NameDropper(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.drop(['FIPS', 'Admin2'], axis=1)

class ProvinceStateTransform(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self,X):
        return X.dropna(subset=['Province_State'])

class LocationImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        imputer=SimpleImputer(strategy="mean")
        X['Lat']=imputer.fit_transform(X[['Lat']])
        X['Long_']=imputer.fit_transform(X[['Long_']])
        return X        
```

```python
df2=pd.read_csv(r"D:\JohnHopkins\COVID-19\csse_covid_19_data\csse_covid_19_daily_reports\01-01-2021.csv")
df2.head(10)
```

```python
dropper=NameDropper()
imp=LocationImputer()
prvState=ProvinceStateTransform()

df2_transform=prvState.fit_transform(imp.fit_transform(dropper.fit_transform(df2)))

df2_transform.columns
df2_transform.columns = [str(col) for col in df2_transform.columns]  # Ensure all column names are strings
df2_transform.dropna(axis='columns', how='all', inplace=True)  # Drop columns where all values are NaN
df2_transform.fillna(value="Default Value", inplace=True)  
df2_transform.head(40)
```

```python
def load_to_sql(df, schema, table_name, connection):
    cursor = connection.cursor()
    # Creating column list for insertion
    cols = ", ".join([f"`{i}`" for i in df.columns.tolist()])

    # Insert DataFrame records one by one.
    for i, row in df.iterrows():
        sql = f"INSERT INTO `{schema}`.`{table_name}` ({cols}) VALUES ({', '.join(['%s']*len(row))})"
        cursor.execute(sql, tuple(row))

    # The connection is not autocommitted by default, so we must commit to save our changes
    connection.commit()
    cursor.close()

# Example usage:
load_to_sql(df2_transform, 'covid_data', 'covid_records', connection)
```
