```python
import pandas as pd

# Creating a DataFrame with default integer row labels
df_default = pd.DataFrame({'A': [10, 20, 30, 40]})
print("DataFrame with Default Integer Labels:\n", df_default)

# Creating a DataFrame with custom non-integer row labels
df_custom_labels = pd.DataFrame({'A': [10, 20, 30, 40]}, index=['one', 'two', 'three', 'four'])
print("\nDataFrame with Custom Labels:\n", df_custom_labels)

# Using label-based slicing on both DataFrames
df_default_sliced_labels = df_default[1:4]
df_custom_sliced_labels = df_custom_labels['two':'four']
df_custom_sliced_label = df_custom_labels[1:]
print("\nLet's Test\n", df_custom_sliced_label)

print("\nLabel-Based Slicing (Default Integer Labels):\n", df_default_sliced_labels)
print("\nLabel-Based Slicing (Custom Labels):\n", df_custom_sliced_labels)

# Using iloc-based slicing on both DataFrames
df_default_sliced_iloc = df_default.iloc[1:4]
df_custom_sliced_iloc = df_custom_labels.iloc[1:4]

print("\niloc-Based Slicing (Default Integer Labels):\n", df_default_sliced_iloc)
print("\niloc-Based Slicing (Custom Labels):\n", df_custom_sliced_iloc)
```

```python
import pandas as pd

# Creating a DataFrame with student data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Grade": ["A", "B", "C", "A", "B"],
    "City": ["New York", "London", "Paris", "Tokyo", "Sydney"]
}
df = pd.DataFrame(data)

# Converting the Grade column to categorical data type
df["Grade"] = df["Grade"].astype("category")

# Displaying the original categorical data
print("Original Categorical Data:")
print(df["Grade"])

# Renaming the categories
df["Grade"] = df["Grade"].cat.rename_categories(["Excellent", "Good", "Average"])

# Reordering the categories and adding missing categories
df["Grade"] = df["Grade"].cat.set_categories(["Average", "Good", "Excellent"])

# Displaying the modified categorical data
print("\nModified Categorical Data:")
print(df["Grade"])
```

```python
import pandas as pd

# Creating a DataFrame with categorical data
data = {
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female"],
    "Age Group": ["Adult", "Adult", "Child", "Senior", "Adult", "Child"]
}

df = pd.DataFrame(data)

# Converting the Gender column to categorical data type
df["Gender"] = df["Gender"].astype("category")

# Displaying the categories and their counts
print(df["Gender"].value_counts())

# Renaming categories
df["Age Group"] = df["Age Group"].cat.rename_categories({"Adult": "18-64", "Child": "0-17", "Senior": "65+"})

# Reordering categories
df["Age Group"] = df["Age Group"].cat.reorder_categories(["0-17", "18-64", "65+"])

# Displaying the modified categorical data
print(df)
```

```python
#Label Encoding in Pandas
import pandas as pd

df = pd.DataFrame({'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small']})

# Define an order for the categorical data
size_order = ['Small', 'Medium', 'Large']

# Convert the 'Size' column to a categorical variable with the defined order
df['Size'] = pd.Categorical(df['Size'], categories=size_order, ordered=True)

# Assign numerical values to the size levels
df['Size_code'] = df['Size'].cat.codes

print(df)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = {
    'Date': pd.date_range(start='2022-01-01', periods=6, freq='D'),
    'Product': ['Widget', 'Gadget']*3,
    'Store': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Sales': np.random.randint(1, 100, size=6),
    'Returns': np.random.randint(0, 5, size=6)
}

df_sales = pd.DataFrame(data)
print(df_sales)
```

```python
#Understanding Groupby

df_aggregated = df_sales.groupby(['Store', 'Product'], as_index=False)['Sales'].sum()
df_aggregated
```

```python
grouped_df = df.groupby(by='grouping_column')['data_column'].sum()
grouped_df = df.groupby(by='grouping_column').agg({'data_column1': 'sum', 'data_column2': 'mean'})
grouped_df = df.groupby(by='grouping_column')['data_column'].agg(lambda x: (x > 10).sum())
df['normalized'] = df.groupby('grouping_column')['data_column'].transform(lambda x: (x - x.mean()) / x.std())
filtered_df = df.groupby('grouping_column').filter(lambda x: x['data_column'].sum() > 100)
```

```python
"""Here, you're grouping the data by 'Store' and then calculating the mean of the 'Sales' for each store. 
The result, single_aggregation, will be a Series where the index is the 'Store' and 
the values are the mean sales for each store."""

single_aggregation = df_sales.groupby(by='Store')['Sales'].mean()
single_aggregation
```

```python
""" This line groups the data by 'Store' and then computes two different aggregations: 
the mean of 'Sales' and the median of 'Returns' for each store. 
The result, multiple_aggregation, is a DataFrame with 'Store' as the index and two columns: 
one for the mean sales and one for the median returns."""

multiple_aggregation = df_sales.groupby(by='Store').agg({'Sales':'mean', 'Returns':'median'})
multiple_aggregation
```

```python
""" If you want to customize the column names in the resulting DataFrame after performing the agg operation, 
you can pass a list of tuples where each tuple contains the desired column name and the aggregation function """

multiple_aggregation = df_sales.groupby(by='Store').agg(
    Mean_Sales=('Sales', 'mean'), 
    Median_Returns=('Returns', 'median')
)
multiple_aggregation 
```

```python
""" Here, you're grouping by 'Product' and then applying a custom lambda function to the 'Returns' column. 
The lambda function calculates the number of times returns were greater than 1 for each product. 
The result is a Series with the count of such instances for each product. """

custom_functions = df_sales.groupby(by='Product')['Returns'].agg(lambda x: (x > 1).sum())
custom_functions
```
