```python
import pandas as pd

dfx=pd.read_csv(r"E:\UoB Workspace\Micro-Courses\Data\archive\ElectricCarData_Norm.csv")
```

```python
dfx.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x="TopSpeed",data=dfx)
```

```python
sns.scatterplot(x="TopSpeed", y="PowerTrain", data=dfx)
```

```python
sns.relplot(x="TopSpeed", y="PowerTrain", data=dfx, kind="scatter")
```

```python
sns.relplot(x="TopSpeed", y="Range", data=dfx, kind="line")
```

```python
sns.relplot(x="TopSpeed", y="Range", data=dfx, kind="line", ci="sd")
```

```python
sns.catplot(x="PowerTrain", data=dfx, kind="count")
```

```python
category_order=['Rear Wheel Drive', 'Front Wheel Drive', 'All Wheel Drive']
sns.catplot(x="PowerTrain", data=dfx, kind="count", order=category_order)
```

```python
sns.catplot(x="Model", y="PriceEuro", data=dfx, kind="bar")
```

```python
sns.catplot(x="Range", y="PriceEuro", data=dfx, kind="box")
```

```python
#Adjusting whiskers
sns.catplot(x="Range", y="PriceEuro", data=dfx, kind="box", whis=2.0)
```

```python
#Setting whiskers to min and max
sns.catplot(x="Range", y="PriceEuro", data=dfx, kind="box", whis=[0,100])
```

```python
#Removing Outliers
sns.catplot(x="Range", y="PriceEuro", data=dfx, kind="box", sym="")
```

```python
#Point Plot
sns.catplot(x="Range", y="Seats", data=dfx, kind="point", capsize=0.2)
```
